```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(tokenized_doc)
```

```python
word2idx = tokenizer.word_index #단어에 인덱스 부여
idx2word = {value: key for key, value in word2idx.items()} #키 벨류 형태
encoded = tokenizer.texts_to_sequences(tokenized_doc)  #인덱스 번호만
```

```python
encoded
```

```python
# 단어 사전의 크기

vocab_size = len(word2idx)+1
print(f'단어 사전의 크기 :  {vocab_size}')
```

```python
# 첫문장
print(encoded[0])
```

```python
#네거티브 샘플링을 위한 모듈 keras에서 제공하는 전처리 도구
from tensorflow.keras.preprocessing.sequence import skipgrams
```

```python
skip_grams = [skipgrams(sample,vocabulary_size=vocab_size,window_size=10) for sample in encoded[:5]]
print(f'전체 샘플 수 : {len(skip_grams)}')
```

```python
pairs,labels = skip_grams[0][0], skip_grams[0][1]
print(f'3 pairs : {pairs[:3]}')
print(f'3 labels : {labels[:3]}')
```

```python
# 첫번째 그룹 샘플에 대해 생긴 pairs와 labels의 개수

print(len(pairs))
print(len(labels))
```

```python
#간단히 시각화
for i in range(5):
    print('({:s} ({:d}), {:s} ({:d})) -> {:d}'.format(
        idx2word[pairs[i][0]], pairs[i][0],
        idx2word[pairs[i][1]], pairs[i][1],
        labels[i]
    ))
```

```python
#트레인 데이터 나누기

train_dataset = [skipgrams(sample,vocabulary_size=vocab_size,window_size=10) for sample in encoded[:1000]]
print(len(train_dataset))
```

```python
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Embedding, Reshape, Activation, Input, Dot
from tensorflow.keras.utils import plot_model
```

```python
#차원을 100으로 줄이겠다.
embedding_dim = 100

#중심 단어를 위한 임베딩 테이블
w_inputs = Input(shape=(1,),dtype='int32')
word_embedding = Embedding(vocab_size,embedding_dim)(w_inputs)


#주변 단어를 위한 임베딩
c_inputs = Input(shape=(1,),dtype='int32')
context_embedding = Embedding(vocab_size,embedding_dim)(c_inputs)
```

```python
dot_product = Dot(axes=2)([word_embedding,context_embedding])

dot_product = Reshape((1,),input_shape=(1,1))(dot_product)

output = Activation('sigmoid')(dot_product)
```

```python
#model 구성하기
model = Model(inputs=[w_inputs,c_inputs],outputs=output)

#모델 구성 보기
model.summary()
```

```python
#손실함수 및 최적화 함수 적용
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
#그림으로 모델 확인하기
plot_model(model,to_file='model.png',show_shapes=True, show_layer_names=True)
```

```python
#학습시키는 것
for epoch in range(10):
    loss = 0
    for _, elem in enumerate(train_dataset):
        first_elem =  np.array(list(zip(*elem[0]))[0],dtype='int32') # *은 튜플을 unpacking하는 것
        second_elem =  np.array(list(zip(*elem[0]))[1],dtype='int32')
        labels = np.array(elem[1],dtype='int32')

        X=[first_elem,second_elem]
        Y=labels

        loss += model.train_on_batch(X,Y)

    print('EPOCH:',epoch+1, 'Loss:',loss)
```

```python
import gensim #모델을 통해 돌렸던 텍스트로 변경해준다.

f=open('vectors.txt','w')
f.write('{} {}\n'.format(vocab_size-1,embedding_dim))

#첫번째 가중치 저장
vectors = model.get_weights()[0]

for word, i in tokenizer.word_index.items():
    f.write('{} {}\n'.format(word, ' '.join(map(str,list(vectors[i,:])))))
f.close()

#품질 확인하기
```
